```
import random
tie=0
user=0
com=0
print("Welcome to Rock Paper Scissors!! \nToday you will be playing against the computer\nALL THE BEST!\n")
comp=["rock","paper","scissor"]
while(1):
    user_inp=input("\nEnter your move\nROCK PAPER OR SCISSORS?\nOr if you want to stop enter STOP\n->")
    a=user_inp.lower()
    if(a not in comp and a != "stop"):
        print("Wrong input Please enter either ROCK or PAPER or SCISSOR")
        continue
    if(a=="stop"):
        print("\nScoreboard\nUSER     TIE     COMPUTER\n  {0}       {1}        {2}".format(user,tie,com))
        break
    r=random.randint(1,3)-1
    b=comp[r]
    print("\nCOMPUTER PLAYS -> ",b)
    print("\n")
    if(a==b):
        print("ROCK...PAPER...SCISSOR...")
        print("\nIt is a tie")
        tie+=1
    elif((a=="rock" and b=="paper") or (b=="rock" and a=="paper")):
        print("ROCK...PAPER...SCISSOR...")
        if(a=="paper"):
            print("\nYOU WON!!")
            user+=1
        else:
            print("\nCOMPUTER WON!!")
            com+=1
    elif((a=="rock" and b=="scissor") or (b=="rock" and a=="scissor")):
        print("ROCK...PAPER...SCISSOR...")
        if(a=="rock"):
            print("\nYOU WON!!")
            user+=1
        else:
            print("\nCOMPUTER WON!!")
            com+=1
    elif((a=="scissor" and b=="paper") or (b=="scissor" and a=="paper")):
        print("ROCK...PAPER...SCISSOR...")
        if(a=="scissor"):
            print("\nYOU WON!!")
            user+=1
        else:
            print("\nCOMPUTER WON!!")
            com+=1
    print("Scoreboard\nUSER     TIE     COMPUTER\n  {0}       {1}        {2}".format(user,tie,com))
    
    



```
