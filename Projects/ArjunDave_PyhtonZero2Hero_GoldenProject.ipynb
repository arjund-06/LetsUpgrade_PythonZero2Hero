```
import sympy

def split_letters(string):
    '''
    THIS FUNCTION IS USED TO SEPERATE THE WORDS IN THE GIVEN STRING
    AND MAKE IT INTO A LIST OF THE ALPHABETS OF THE STRING 
    '''
    str_lst=[]
    for i in range(len(string)):
        str_lst.append(string[i])
    return str_lst

def ASCII_lst(str_lst):
    '''
    THIS FUNCION CONVERTS THE ALPHABETS IN THE LIST OF STRINGS INTO 
    THEIR RESPECTIVE ASCII NUMBERS AND PUTS IT IN ANOTHER LIST OF ASCII NUMBERS
    REPRESENTING EACH ALPHABET OF THE STRING
    '''
    ascii_lst= list(map(lambda x: ord(x),str_lst))
    return ascii_lst 


def prime_ascii(numero):
    '''
    THIS FUNCTION FINDS THE CLOSEST PRIME NUMBER TO THE SPECIFIC NUMBER GIVEN AS 
    ARGUMENT. IF THERE ARE 2 PRIME NUMBERS EQUALLY APART FROM THE SPECIFIED 
    NUMBER IT CHOOSES THE SMALLER NUMBER
    '''
    for n in range(numero,0,-1):
        if(sympy.isprime(n)==True):
            n_side=numero-n
            break
    for p in range(numero,150):
        if(sympy.isprime(p)==True):
            p_side=p-numero
            break
    if (p_side==n_side):
        return n
    elif (p_side<n_side):
        return p
    else:
        return n        

def nearest_primeno(number):
    '''
    THIS FUNCTION LINKS BOTH THE MAIN CODE AND THE 'prime_ascii(numero)' FUNCTION
    BY TAKING IN THE NUMBER FROM THE MAIN CODE AND USING THE FUNCTION RETURNS THE 
    CLOSEST PRIME NUMBER
    '''
    if sympy.isprime(number)==True:
        return number
    else:
        pa=prime_ascii(number)
        return pa

# THE MAIN CODE STARTS HERE

test_case=int(input("Enter the number of testcases:\n->"))
while(1):
    if (test_case > 500 or test_case < 1) and test_case >= 0:
        print("Please Enter a value between 1 and 500")
        test_case=int(input("Enter the number of testcases:\n->"))
    elif test_case < 0:
        print("Please enter a positive value")
        test_case=int(input("Enter the number of testcases:\n->"))
    else:
        break 
iterations=1
while(iterations<=test_case): 
    str_length=int(input("Enter length of string\n->"))
    while(1):
        if (str_length > 500 or str_length < 1) and str_length >= 0:
            print("Please enter a length between 1 and 500")
            str_length=int(input("Enter length of string\n->"))
        elif str_length < 0:
            print("Please enter a positive value for length")
            str_length=int(input("Enter length of string\n->"))
        else:
            break
    string=input("Enter the string:\n->")
    while(len(string) != str_length):
        print("Length of string is not as specified above")
        string=input("Enter the string:")
      
    str_lst = split_letters(string)
    ascii_lst = ASCII_lst(str_lst)
    prime_lst = list(map(nearest_primeno,ascii_lst)) #GETS ALL THE NEAREST PRIME
                                                     #PRIME NUMBERS OF EVERY NUMBER
                                                     #IN THE LIST
    final_lst = list(map(lambda x: chr(x),prime_lst)) #CHANGES ALL THE ASCII NUMBERS
                                                      #IN THE LIST INTO THEIR 
                                                      #RESPECTIVE ALPHABETS
    final_string = ""
    for i in final_lst:
        final_string = final_string + i
    print("____________________\nPanda's Magical Word:")
    print(final_string,"\n")
    iterations = iterations + 1  
    
```
